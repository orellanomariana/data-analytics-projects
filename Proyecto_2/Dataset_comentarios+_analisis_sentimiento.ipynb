```python
import re

def preprocess_text(text):
    # Función de preprocesamiento
    text = text.lower()
    text = re.sub(r'\s+', ' ', text)
    text = re.sub(r'[^\w\s]', '', text)
    return text
df['Comentario'] = df['Comentario'].apply(preprocess_text)


```

```python
from transformers import pipeline

# Cargar un modelo preentrenado para análisis de sentimientos en español
sentiment_pipeline = pipeline("sentiment-analysis", model="nlptown/bert-base-multilingual-uncased-sentiment")

def get_sentiment(text):
    result = sentiment_pipeline(text)[0]
    label = result['label']

    # Clasificar según el label
    if label in ['5 stars', '4 stars']:
        return 'Positivo'
    elif label in ['3 stars']:
        return 'Neutral'
    else:
        return 'Negativo'

# Aplicar la función a la columna "Comentario"
df['Sentimiento'] = df['Comentario'].apply(get_sentiment)

# Mostrar los primeros resultados para verificar
df.head(10)


```
